# 4. 스케일링

6-1. col-min(col)로 양수만들기

```python
# Train/Test 데이터의 수치형 변수들에 대해 scaling 진행
train_col=train_c.drop(['OC','sido','ownerChange','instkind'],axis=1)
test_col=test_c.drop(['OC','sido','ownerChange','instkind'],axis=1)
```

```python
# 로그 변환 이전, Train 데이터 칼럼의 음수 값을 제거하기 위한 과정
for i in train_col.columns:
  train_col[i] -= min(train_col[i])
```

```python
# 로그 변환 이전, Test 데이터 칼럼의 음수 값을 제거하기 위한 과정
for i in test_col.columns:
  test_col[i] -= min(test_col[i])
```

6-2. log scaling

```python
train_num_scaled_log=np.log1p(train_col)
```

```python
test_num_scaled_log=np.log1p(test_col)
```

6-3. one-hot encoding

```python
#범주형 변수 Dummy Variables
sum_str=sum[['OC','sido','ownerChange','instkind']]
sum_str=pd.get_dummies(sum_str)
```

```python
# train, test 분리
train_str = sum_str.iloc[0:301,:]
test_str = sum_str.iloc[301:,:]
```

```python
# train
X_train=pd.concat([train_num_scaled_log,train_str], axis=1,ignore_index=False)
X_test=pd.concat([test_num_scaled_log,test_str], axis=1,ignore_index=False)
```

```python
#변수 나누기
X=X_train.drop(['OC_open','OC_ close'],axis=1)
y=X_train['OC_open']
```

```python
X_test=X_test.drop(['OC_ close'],axis=1)
```

```python
# import pycaret
from pycaret.classification import *
```

model setup

```python
model = setup(
    data = X_train.drop('OC_ close',axis=1),
    target = "OC_open",
    fold = 15
)
```

모델 비교

```python
top5_model = compare_models(
               round=4,
               sort="Accuracy",
               n_select = 5)
```

```python
top5_model
```

```python
del top5_model[1] #dummyClassifier은 뺌
```

KNeighborClassifier, SGDClassifier, ExtraTreesClassifier, CatBoostClassifier 최종선택

모델튜닝

```python
tuned_top5 = [tune_model(i,choose_better = True) for i in top5_model]  
```

```python
tuned_top5
```

Staking

```python
blend5_stack = stack_models(estimator_list=tuned_top5)
```

```python
final_model = finalize_model(blend5_stack)
```

```python
prediction = predict_model(final_model, data = X_test) 
```
